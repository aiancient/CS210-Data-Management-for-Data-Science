```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sb

#PREPROCESSING/ANALYSIS
#part1
mpgs = pd.read_csv("EuCitiesTemperatures.csv")
mpgs['latitude'] = mpgs.groupby('country')['latitude'].transform(lambda x: x.fillna(x.mean())).round(2)
mpgs['longitude'] = mpgs.groupby('country')['longitude'].transform(lambda x: x.fillna(x.mean())).round(2)
mpgs
```

```python
#part2
subsetOfCities = mpgs[((40 <= mpgs['latitude']) & (mpgs['latitude'] <= 60)) & ((15 <= mpgs['longitude']) & (mpgs['longitude'] <= 30))]
country_counts = subsetOfCities['country'].value_counts()
country_counts[country_counts == country_counts.max()]
```

```python
#part3
mpgs['temperature'] = mpgs.groupby(['EU','coastline'])['temperature'].transform(lambda x: x.fillna(x.mean())).round(2)
mpgs
```

```python
#VISUALIZATION
#part1
mpgsEUCoastlineData = mpgs.groupby(['EU','coastline']).count()['city'].to_frame('city_count')
mpgsEUCoastlineData
plt.figure(figsize=(8,5))
plt.title('Num of Cities Belonging to EU & Coastline',fontsize='14')
plt.xlabel('EU,coastline')
plt.ylabel('City Count')
mpgsEUCoastlineData['city_count'].plot(kind = 'bar')
```

```python
#part2
plt.figure(figsize=(8,5))
plt.title("Longitude-Latitude Map like plot")
sb.scatterplot(x=mpgs['longitude'],y=mpgs['latitude'],hue=mpgs['country'])
plt.show()   
```

```python
#part3
plt.figure(figsize=(8,5))
plt.title('Number of Countries belonging to the Population group')
plt.xlabel('Population group')
plt.ylabel('Frequency')
plt.hist(mpgs['population'],bins=5,edgecolor='black')
```
